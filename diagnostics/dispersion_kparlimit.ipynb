```python
import dielectric
from scipy.optimize import fsolve

th = 1.
kx = 0.1
ky = 0.1
kpar = np.linspace(0.001, 0.00001, 1001)
kapn = 0.00
kapt = 0.00
alpha = 0.0
memi = 1./1836.

guess = np.array([[-0.7,0.0],[0.7,0.0],[1.99,0.0],[2.999,0.0]])
om = np.zeros((guess.shape[0],len(kpar)))
gm = np.zeros((guess.shape[0],len(kpar)))

for i in range(guess.shape[0]):
    for j in range(len(kpar)):
        om[i,j], gm[i,j] = fsolve(dielectric.DKEdielectricFunc, guess[i,:], args = (th, kx, ky, kpar[j], kapn, kapt, alpha, memi))
        guess[i,:] = np.array([om[i,j],gm[i,j]])
        if j % 10 == 0:
          print('kpar = {0:.6f}, omega = {1:.6f}, gamma = {2:.6f}'.format(kpar[j],om[i,j],gm[i,j]))
    print()
```
